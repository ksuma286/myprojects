```python
import pandas as pd
```

```python
df=pd.read_csv('netflix_titles.csv')
```

```python
df['director'].fillna('Missing data',inplace=True)
df['cast'].fillna('Missing data',inplace=True)
df['country'].fillna('Missing data',inplace=True)
df['date_added'].fillna('Missing data',inplace=True)
df['rating'].fillna('Missing data',inplace=True)
df['duration'].fillna('Missing data',inplace=True)
```

Wordcloud is visualization of the frequency of words in a text where the size of the word represents its frequency.



Making a word cloud out of description of TV Shows

```python
df_tv_show=df[df['type']=='TV Show']
```

```python
text=(' ').join(i.replace('.',' ').replace(',',' ').replace(';',' ') for i in df_tv_show['description'])
```

```python
!pip install wordcloud
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
word_cloud = WordCloud(
width=3000,
height=2000,
random_state=1,
background_color="black",
colormap="Pastel1",
collocations=False,
stopwords=STOPWORDS,
).generate (text)
```

width/height : To adjust the height and width of the wordcloud
random_state : To recreate the same plot every time we run the function. The random_state parameter has to be an integer value.
background_color : To set a background_color. The default value for this parameter is 'black'.
colormap : To set up the color theme for the words.
stopwords : To set the list of words that needs to be eliminated.This list can include words like this, that, is, was, the, etc. If this parameter is set to None, then function will consider a built-in list of STOPWORDS

```python
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```

Making a word cloud out of descprtion of Movies

```python
df_movie=df[df['type']=='Movie']
```

```python
text=(' ').join(i.replace('.',' ').replace(',',' ').replace(';',' ') for i in df_movie['description'])
```

```python
word_cloud = WordCloud(
width=3000,
height=2000,
random_state=1,
background_color="white",
colormap="twilight_shifted",
collocations=False,
stopwords=STOPWORDS,
).generate (text)
```

```python
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
