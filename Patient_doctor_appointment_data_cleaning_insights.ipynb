# Loading the libraries needed

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplolib inline
```

```python
dataf=pd.read_csv('noshowappointments-kagglev2-may-2016.csv')
```

Accoriding to the google sheet information in which a description has been provided for this particluar dataset,No means the person showed up for his/her appointment and Yes means they did not show up for their appointment.

In the dataframe "YES" we take all the records who did not show up for their appointment which is indicated as 'Yes' in dataset.

In the dataframe "NO" we took all the records who showed up for their appointment which is indicated by 'No' in dataset

I updated the values of No-show column to 'Yes' for patients who showed up for thier appointment and 'No' for patients who did not show up for their appointment.

```python
YES=dataf[dataf['No-show']=='Yes']
NO=dataf[dataf['No-show']=='No']
```

```python
YES.head(2)
```

```python
NO.head(2)
```

```python
YES.replace('Yes','No',inplace=True)
NO.replace('No','Yes',inplace=True)
```

```python
YES.head(2)
```

```python
NO.head(2)
```

```python
df=NO.append(YES)
```

```python
df.head(2)
```

# Changed datatypes of the fields as per requirement

```python
names=['Scholarship','Hipertension','Diabetes','Alcoholism','Handcap','SMS_received']
for i in names:
    df[i]=df[i].astype('object')
```

```python
df.info()
```

```python
df.describe(include=object)
```

In the scholarship,Hipertension,Diabetes,Alcoholism,Handcap,SMS_received columns 0 means no and 1 means yes is the understanding

There seems to be some extra values in Handcap data which needs to be looked into


```python
df['PatientId']=df['PatientId'].astype('object')
df['AppointmentID']=df['AppointmentID'].astype('object')
```

# Looking into Scheduled and Appointment dates

```python
df['ScheduledDay']=pd.to_datetime(df['ScheduledDay']).dt.date
```

```python
df['AppointmentDay']=pd.to_datetime(df['AppointmentDay']).dt.date
```

# It does not make sense if the scheduling date falls after the appointment date so,deleting such records from dataset

```python
df[df['ScheduledDay']>df['AppointmentDay']]
```

```python
d=df[df['ScheduledDay']>df['AppointmentDay']]
```

```python
df.drop(d.index,inplace=True)
```

```python
df[df['ScheduledDay']>df['AppointmentDay']]
```

Yes
