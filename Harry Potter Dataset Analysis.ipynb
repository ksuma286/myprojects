```python
df_c_dc
```

```python
df_dummy=df1[['Chapter ID','Movie ID']]
```

```python
df_dummy_1=df3[['Movie ID','Movie Title']]
```

```python
df_dummy=pd.merge(df_dummy,df_dummy_1,on='Movie ID',how='left')
```

```python
df_dummy
```

```python
df_dummy=pd.merge(df_dummy,df_c_dc,on='Chapter ID')
```

```python
df_dummy
```

```python
df_dummy_2=df2[['Character ID','Character Name']]
```

```python
df_dummy=pd.merge(df_dummy,df_dummy_2,on='Character ID')
```

```python
df_dummy
```

```python
top_characters_with_most_dialogues=df_dummy.groupby('Character Name')['Number of Dialogues'].sum().sort_values(ascending=False).reset_index()
```

```python
top_characters_with_most_dialogues.head()
```

```python
top_5_characters_with_most_dialogues=top_characters_with_most_dialogues.loc[0:4,'Character Name']
```

```python
df_dummy=df_dummy[df_dummy['Character Name'].isin(top_5_characters_with_most_dialogues['Character Name'])]
```

```python
df_dummy=df_dummy.groupby(['Movie Title','Character Name'])['Number of Dialogues'].sum().reset_index()
```

Which spells are used the most across the franchise? Do characters have a favorite spell?

```python
text=(' ').join(i for i in df4['Dialogue'])#appending all dialogues into a single sentence
```

```python
d={'1':'0'}
for i in df6['Incantation']:
    x=text.count(i)#counting number of time a spell has occured in entire movie franchise 
    d[i]=x#appending spell and its count to a dictionary d
    
```

```python
d
```

```python
df_spell=pd.DataFrame(d.items(),columns=['Spell','Count'])
```

```python
df_spell
```

```python
df_spell.drop(0,axis=0,inplace=True)
```

```python
df_spell
```

```python
df_spell=df_spell.sort_values(by='Count',ascending=False)
```
